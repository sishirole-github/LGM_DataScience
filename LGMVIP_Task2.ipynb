**Importing required libraries**

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

 **LOADING THE IMAGE**



```python
image=cv2.imread("/content/puppy-1903313__340.jpg",cv2.IMREAD_UNCHANGED)
```

```python
cv2_imshow(image)
```

**CONVERTING THE IMAGE INTO GRAY SCALE IMAGE**

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

**Converting image into gray scale will give us black and white pixel which is required for converting the iamge into pencil sketch**

**INVERTING THE IMAGE**

```python
image_invert = cv2.bitwise_not(image_gray)
cv2_imshow(image_invert)
#invert_image=255-grey_image
```

**Used bit_not function to find the brighter and the lighter part in the image so that we can find the edge to create the pencil sketch**

**APPLYING SMOOTHING TO THE INVERTED IMAGE**

```python
image_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

**By Smoothing the image we cxan reduce  the noice and find the lines to create a sketch** 

**FUNCTION FOR GEETING THE PENCIL SKETCH**

```python
def dogV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dogV2(image_gray, image_smoothing)
cv2_imshow(final_img)
```
